```python
import pandas as pd
```

```python
annotate = pd.read_csv("userC-annotate.csv", header=None, index_col=0)
annotate.columns = ["name", "rel"]
annotate = annotate.set_index("name")
```

```python
rel_dict = annotate["rel"].to_dict()
```

```python
rel_dict
```

```python
df = pd.read_csv("userC-0click.csv", header=None, index_col=0)
```

```python
df.columns = ["doc_id", "name", "rel"]
```

```python
df["rel"] = df["name"].map(rel_dict, na_action='ignore')
```

```python
df = df.drop("doc_id", axis=1)
```

```python
df.to_csv("userC-0click.csv", header=False)
```
