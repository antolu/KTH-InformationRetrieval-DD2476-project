```python
import pandas as pd
```

```python
annotate = pd.read_csv("userB-annotate.csv", header=None, index_col=0)
annotate.columns = ["name", "rel"]
annotate = annotate.set_index("name")
```

```python
rel_dict = annotate["rel"].to_dict()
```

```python
df0 = pd.read_csv("userB-0click.csv", header=None, index_col=0)
```

```python
df0.columns = ["doc_id", "name"]
```

```python
df0["rel"] = df0["name"].map(rel_dict, na_action='ignore')
```

```python
df0
```

```python
df0.to_csv("userB-0click.csv", header=False)
```
